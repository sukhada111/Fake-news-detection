```python
unknown_pubs = []
for index,row in enumerate(true.text.values):
    try:
        record = row.split(" -", maxsplit=1)
        record[1]
        assert(len(record[0]) < 260)
    except:
        unknown_pubs.append(index)
```

```python
true.iloc[unknown_pubs].text
```

```python
pub_info = []
news_text = []
for index,row in enumerate(true.text.values):
    if index in unknown_pubs:
        news_text.append(row)        
        pub_info.append("Unknown")
        continue
    record = row.split(" -", maxsplit=1)
    pub_info.append(record[0])
    news_text.append(record[1])
```

```python
true["publisher"] = pub_info
true["news_text"] = news_text

del pub_info, news_text, record, unknown_pubs
```

```python
nltk.download('punkt')
X = []
stop_words = set(nltk.corpus.stopwords.words("english"))
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
for par in df["text"].values:
    tmp = []
    sentences = nltk.sent_tokenize(par)
    for sent in sentences:
        sent = sent.lower()
        tokens = tokenizer.tokenize(sent)
        filtered_words = [w.strip() for w in tokens if w not in stop_words and len(w) > 1]
        tmp.extend(filtered_words)
    X.append(tmp)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=111)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('bow', CountVectorizer()),
    ('tfidf', TfidfTransformer()),  
    ('classifier', MultinomialNB()),
])
```

```python
X=df['text']
y=df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=101)
```

```python
pipeline.fit(X_train,y_train)
```

```python
pred=pipeline.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
